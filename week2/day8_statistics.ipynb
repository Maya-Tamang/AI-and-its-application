#Measure of central tendency

Note! We can also use module statistics instead of Numpy and pandas dor mathamtical statistics in python

```python
import pandas as pd
data=pd.Series([3,5,7,9,2,3,5,3,10,13,16,18])

#Calculating mean
mean=data.mean()
print(f"The mean of data is:{mean}")


#Calculating median
median=data.median()
print(f"The median of data is:{median}")

mode=data.mode()

print("The mode of data is:",mode[0])
```

#Measure of Variability

#Range

```python
#calculating range
data=pd.Series([3,5,7,9,2,3,5,3,10,13,16,18])
range=data.max()-data.min()
print(range)
```

# Task 4: Create a Pandas Series with values [7, 14, 21, 28]. Calculate and print the range.

```python
data = pd.Series([7, 14, 21, 28])

range_value = data.max() - data.min()
print("Range:", range_value)
```

#Variance and standard deviation

```python
#Calculating variance and standard deviation
data=pd.Series([3,5,7,9,2,3,5,3,10,13,16,18])
variance=data.var()
standard_deviation=data.std()

print(variance)
print(standard_deviation)
```

# Task 5: Create a Pandas Series with values [5, 10, 15, 20]. Calculate and print the variance and standard deviation.

```python
data = pd.Series([5, 10, 15, 20])

variance = data.var()
std_dev = data.std()

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

#Data with Matplotlib

```python
#visualizing data with Matplotlib
import matplotlib.pyplot as plt

data=pd.DataFrame({
    "scores":[45,67,34,78,25,89,12,67,90,200,150,250]
})
data.boxplot(column="scores")
plt.show()
```

# Task 6: Create a DataFrame with a column ”Values” containing [10, 20, 15, 30,
# 25]. Create a box plot, add a title and y-label, and save it as valuesboxplot
png.

```python
df = pd.DataFrame({'Values': [10, 20, 15, 30, 25]})

plt.boxplot(df['Values'])
plt.title('Box Plot of Values')
plt.ylabel('Values')

plt.savefig('valuesboxplot.png')
plt.show()
```
